# NY CitiBike Analysis: Additional Plots

```python
# Import libraries
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
from datetime import datetime as dt
import seaborn as sns
```

```python
# Load data
df = pd.read_csv('new_york_data.csv', index_col = 0)
```

```python
df.dtypes
```

```python
# Set theme for all plots
sns.set_theme(style = 'whitegrid', palette = 'Greens_r')
```

## Most Popular Starting Stations

```python
# Create top20 df of most popular starting stations
df['value'] = 1
```

```python
df_t = df.groupby('start_station_name')['value'].count().reset_index()
```

```python
top20 = df_t.nlargest(20, 'value')
```

```python
top20
```

```python
# Bar plot 
plt.figure(figsize = (10, 5))
with sns.color_palette('Greens_r', n_colors = 20):
    sns.barplot(data = top20, x = 'value', y = 'start_station_name').set_title('Most Popular Start Stations')
plt.show()
```

#### The color palette chosen when we set the theme above makes sense for this bar plot. We did need to redefine the color and specify that we need 20 colors for the 20 different stations that would be plotted. We had to take this extra step because the original color palette contains only 12 colors.

## Monthly Rides vs Weather

```python
# Add bike_rides_daily column
df['date'] = pd.to_datetime(df['date'], format = '%Y-%m-%d')
```

```python
df_group = pd.DataFrame(df.groupby(['date'])['ride_id'].count()).reset_index()
```

```python
df_group.head()
```

```python
df_group.rename(columns = {'ride_id': 'bike_rides_daily'}, inplace = True)
df = df.merge(df_group, on = 'date', how = 'outer', indicator = 'merge_flag')
print(df['merge_flag'].value_counts(dropna = False))
print("Shape of January 1st is", df[df['date'] == '2018-01-01'].shape) # Check 
print("Shape of January 2nd is", df[df['date'] == '2018-01-02'].shape) # Second check 
```

```python
# Remove merge_flag column
df.drop(columns = {'merge_flag'}, inplace = True)
```

```python
# Create temp df where the date is the index
df_temp = df.set_index('date')
```

```python
# Line plot
fig, ax = plt.subplots(figsize = (10, 5))
sns.lineplot(data = df_temp['bike_rides_daily_x'], color = "g")
ax.set_xlabel('Year 2022', fontsize = 14)
ax.set_ylabel('Bike Rides Daily', color = 'green', fontsize = 14)

ax2 = ax.twinx()
sns.lineplot(data = df_temp['avgTemp'], color = 'r', ax = ax2)
ax2.set_ylabel('Average Temperature', color = 'red', fontsize = 14)

plt.title('Rides and Temperature New York 2022', fontsize = 18)
plt.show()
```

## Ride Weather by Member Type

```python
# Box plot
plt.figure(figsize = (10, 5))
sns.boxplot(x = 'member_casual', y = 'avgTemp', data = df)
plt.show()
```

#### Based on the above box plot, the data shows that members rent bikes on days with a 1.5 degrees Celcius lower average temperature than days that non-members, or casual riders, rent bikes. Additionally, the minimum temperature that a member will rent a bike is 5.3 degrees Celcius lower than that for a non-member. This could be indicative of the fact that members are more likely to use the bikes year round, perhaps for a daily commute or just as general transportation, while non-members may be using the bikes as a one off. Perhaps they are in the city on vacation and would only choose to use the bikes if the weather is nice outside.

## Ride Weather by Bike Type

```python
# Look at classic and electric bikes
df_sub = df[df['rideable_type'] != 'docked_bike']
```

```python
# FacetGrid
grid = sns.FacetGrid(df_sub, col = 'rideable_type')
grid.map(sns.histplot, 'avgTemp', bins = 15)
plt.show()
```

#### Based on the above histograms, the data shows that classic bikes are rented more often when the temperature is warmer. Electric bikes are rented slightly more often when the temperature is warmer, however relative to classic bikes, electric bike rentals do not change much with the temperature.

#### Additionally, the data shows that classic bikes are rented over 2.5 times more often than electric bikes. This limited electric bike availablity helps to explain why their rental does not change much with the weather, as they may simply not be an option most of the time.
